```python
import pandas as pd
import numpy as np
```

```python
#Antidepresivos
```

```python
df_sucio = pd.read_csv('consumo_ansioliticos_ocde.csv')
```

```python
antidepresivos = ['N06A-Antidepressants']
```

```python
df_seleccion = df_sucio[df_sucio['Variable'].isin(antidepresivos)]
```

```python
df_seleccion
```

```python
df_media_ocde = df_seleccion.groupby(by = 'Country').median()
df_media_ocde['Value'].median()
```

```python
pais = ['Spain']
```

```python
df_media_españa = df_seleccion[df_seleccion['Country'].isin(pais)]
df_media_españa = df_media_españa['Value'].median()
```

```python
df_media_españa
```

```python
df_evolucion_ocde = df_seleccion.groupby(by = 'Year').median()
df_evolucion_ocde
```

```python
df_media_españa = df_seleccion[df_seleccion['Country'].isin(pais)]
df_media_españa
```
