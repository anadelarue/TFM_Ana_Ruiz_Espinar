```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_rows", None)
```

```python
df_sucio = pd.read_csv('asalariados_por_tipo_de_contrato_o_relacion_laboral,_sexo_y_grupo_de_edad.csv', encoding = 'latin1', sep = ';')
```

```python
df_sucio['Tipo de contrato o relación laboral'].value_counts()
```

```python
año = [2020]
```

```python
df_sucio = df_sucio[df_sucio['Periodo'].isin(año)]
```

```python
df_sucio
```

```python
df_sucio['Total'] = df_sucio['Total'].map(lambda x: x.replace(',', '.')).astype(float)
```

```python
df_sucio['Total'] = df_sucio['Total'].map(lambda x: x * 10000)
```

```python
df_sucio['Total'] = df_sucio['Total'].astype(int)
```

```python
df_sucio = df_sucio.groupby(by = ['Tipo de contrato o relación laboral', 'Sexo'], as_index = False).median()
```

```python
df_sucio
```

```python
hombres = ['Hombres']
```

```python
mujeres = ['Mujeres']
```

```python
df_hombres = df_sucio[df_sucio['Sexo'].isin(hombres)]
```

```python
df_mujeres = df_sucio[df_sucio['Sexo'].isin(mujeres)]
```

```python
df_hombres_indefinido = df_hombres[0:3]
```

```python
df_hombres_indefinido['Porcentajes'] = df_hombres_indefinido['Total'].map(lambda x: x * 100 / 1524000)
```

```python
df_hombres_indefinido
```

```python
df_hombres_temporal = df_hombres[-10:]
df_hombres_temporal
```

```python
df_hombres_temporal['Porcentajes'] = df_hombres_temporal['Total'].map(lambda x: x * 100 / 2400000)
```

```python
df_hombres_temporal = df_hombres_temporal.sort_values(by = ['Porcentajes'], ascending = False).head(6).drop(df_hombres_temporal.index[0])
df_hombres_temporal
```

```python
df_mujeres
```

```python
df_mujeres_indefinido = df_mujeres[0:3]
```

```python
df_mujeres_indefinido['Porcentajes'] = df_mujeres_indefinido['Total'].map(lambda x: x * 100 / 1094000)
```

```python
df_mujeres_indefinido
```

```python
df_mujeres_temporal = df_mujeres[-10:]
df_mujeres_temporal
```

```python
df_mujeres_temporal['Porcentajes'] = df_mujeres_temporal['Total'].map(lambda x: x * 100 / 2276000)
```

```python
df_mujeres_temporal = df_mujeres_temporal.sort_values(by = ['Porcentajes'], ascending = False).head(6).drop(df_mujeres_temporal.index[0])
df_mujeres_temporal
```

```python
df_temporal_total = pd.concat([df_hombres_temporal, df_mujeres_temporal])
df_temporal_total
```

```python
temporalidad = ['De duración indefinida: Total', 'Temporal: Total']
```

```python
df_sucio_resumen = df_sucio[df_sucio['Tipo de contrato o relación laboral'].isin(temporalidad)]
df_sucio_resumen
```

```python
porcentaje_hombres_indefinido = df_sucio_resumen.iloc[0,3] * 100 / (df_sucio_resumen.iloc[0, 3] + df_sucio_resumen.iloc[2,3])
porcentaje_hombres_indefinido
```

```python
porcentaje_hombres_temporal = df_sucio_resumen.iloc[2,3] * 100 / (df_sucio_resumen.iloc[0, 3] + df_sucio_resumen.iloc[2,3])
porcentaje_hombres_temporal
```

```python
porcentaje_mujeres_indefinido = df_sucio_resumen.iloc[1,3] * 100 / (df_sucio_resumen.iloc[0, 3] + df_sucio_resumen.iloc[2,3])
porcentaje_mujeres_indefinido
```

```python
porcentaje_mujeres_temporal = 100 - porcentaje_mujeres_indefinido
porcentaje_mujeres_temporal
```
