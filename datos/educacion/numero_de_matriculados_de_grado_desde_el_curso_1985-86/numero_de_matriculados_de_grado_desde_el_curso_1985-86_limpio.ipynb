```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('numero_de_matriculados_de_grado_desde_el_curso_1985-86.csv', header = [1])
```

```python
df_seleccionado = df[0:6]
```

```python
df_seleccionado_dos = df_seleccionado.iloc[:, 0:37]
```

```python
df_seleccionado_dos
```

```python
df_limpio = df_seleccionado_dos.drop(df.index[[0,1,5]]).rename(columns={"Unnamed: 0":"AÃ±o"})
df_limpio
```

```python
to_numeric = pd.to_numeric(df_limpio.iloc[0, 18])
```

```python
evolucion_ambos_sexos = (df_limpio.iloc[0, 36] * 100 / to_numeric) - 100
evolucion_ambos_sexos.round(decimals = 2)
```

```python
evolucion_hombres = (df_limpio.iloc[1, 36] * 100 / pd.to_numeric(df_limpio.iloc[1, 18])) - 100
evolucion_hombres.round(decimals = 2)
```

```python
evolucion_mujeres = (df_limpio.iloc[2, 36] * 100 / pd.to_numeric(df_limpio.iloc[2, 18])) - 100
evolucion_mujeres.round(decimals = 2)
```
