```python
import pandas as pd
import numpy as np
```

```python
df_sucio = pd.read_csv('saldo_migratorio_interprovincial_por_semestre,_provincia,_sexo_y_generacion_(edad_a_31_de_diciembre).csv', encoding = 'latin1', on_bad_lines='skip', sep=';')
```

```python
df_sucio
```

```python
años_si = ['2020S2', '2020S1']
```

```python
df_sucio_semestres = df_sucio[df_sucio['Periodo'].isin(años_si)]
```

```python
df_sucio_semestres
```

```python
df_casi_limpio = df_sucio_semestres.groupby(by = 'Provincias', as_index=False).sum().sort_values(by = 'Total', ascending=False)
df_casi_limpio
```

```python
los_no = ['38 Santa Cruz de Tenerife', '35 Palmas, Las']
```

```python
df_casi_limpio = df_casi_limpio[df_casi_limpio['Provincias'].isin(los_no) == False]
```

```python
provincias = ['Madrid', 'Girona', 'Guadalajara', 'Tarragona', 'A Coruña', 'Toledo', 'Illes Balears', 'Castelló/Castellón', 'Málaga', 'València/Valencia', 'Cantabria', 'Lleida', 'Huesca', 'Navarra', 'Pontevedra', 'Gipuzkoa', 'Bizkaia', 'Teruel', 'Sevilla', 'Alacant/Alicante', 'Soria', 'Zaragoza', 'Araba/Álava', 'Ávila', 'Segovia', 'Ourense', 'Burgos', 'Cuenca', 'La Rioja', 'Lugo', 'Zamora', 'Valladolid', 'Asturias', 'Murcia', 'Granada', 'Palencia', 'Cádiz', 'Huelva', 'Salamanca', 'Albacete', 'Ceuta', 'Melilla', 'Almería', 'Ciudad Real', 'Cáceres', 'León', 'Badajoz', 'Córdoba', 'Jaén', 'Barcelona']
```

```python
df_casi_limpio['Provincias_bien'] = provincias
```

```python
df_casi_limpio = df_casi_limpio.drop('Provincias', axis = 1)
```

```python
df_casi_limpio
```

```python
df_casi_limpio = df_casi_limpio.rename(columns={'Provincias_bien': 'Provincias'})
df_casi_limpio = df_casi_limpio.reset_index()
```

```python
df_casi_limpio = df_casi_limpio.drop('index', axis = 1)
```

```python
df_casi_limpio.to_csv('saldo_migratorio_interprovincial_por_semestre,_provincia,_sexo_y_generacion_(edad_a_31_de_diciembre)_limpio.csv', index = False)
```
