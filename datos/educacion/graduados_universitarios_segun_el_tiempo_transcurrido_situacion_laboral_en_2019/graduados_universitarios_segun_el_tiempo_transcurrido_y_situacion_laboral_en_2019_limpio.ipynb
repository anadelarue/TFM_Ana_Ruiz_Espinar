```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_rows", None)
```

```python
df_sucio = pd.read_csv('graduados_universitarios_segun_el_tiempo_transcurrido_desde_que_acabaron_la_carrera_hasta_que_encontraron_su_primer_empleo_por_sexo_y_situacion_laboral_en_2019.csv', encoding = 'latin1', sep = ';')
```

```python
df_sucio
```

```python
lo_que_si = ['Trabajando']
```

```python
lo_que_no = ['Continuó al menos 6 meses más en el trabajo que tenía mientras estudiaba']
```

```python
df_sucio = df_sucio[df_sucio['Situación laboral en 2019'].isin(lo_que_si)]
```

```python
df_sucio = df_sucio[df_sucio['Tiempo transcurrido desde que acabaron la carrera hasta su primer empleo'].isin(lo_que_no) == False]
```

```python
df_sucio = df_sucio.groupby(by = 'Tiempo transcurrido desde que acabaron la carrera hasta su primer empleo').mean()
```
