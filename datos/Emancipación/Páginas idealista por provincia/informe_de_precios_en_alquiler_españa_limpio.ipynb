```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None)
```

```python
pd.options.mode.chained_assignment = None
```

```python
html = pd.read_html('Informe de precios en alquiler en España — idealista.html')
```

```python
df_sucio = html[0]
```

```python
df_sucio.dtypes
```

```python
discard_provincia = ["provincia"]
```

```python
discard_ccaa = ('España', 'Andalucía', 'Aragón', 'Oviedo', 'Palma de Mallorca', 'Canarias', 'Las Palmas', 'Santander', 'Castilla y León', 'Castilla-La Mancha', 'Cataluña', 'Comunitat Valenciana', 'Castellón de la Plana / Castello de la Plana', 'Euskadi', 'Donostia-San Sebastián', 'Vitoria-Gasteiz', 'Extremadura', 'Galicia', 'Logroño', 'Madrid Comunidad', 'Murcia Región', 'Pamplona/Iruña')
```

```python
df_no_provincia = df_sucio[~df_sucio['Localización'].str.contains('|'.join(discard_provincia))]
```

```python
df_no_ccaa = df_no_provincia[df_no_provincia['Localización'].isin(discard_ccaa) == False]
df_no_ccaa
```

```python
df_no_ccaa['Precio m2 abr 2022'] = df_no_ccaa['Precio m2 abr 2022'].map(lambda x: str(x)[:-4])
df_no_ccaa
```

```python
df_punto_coma = df_no_ccaa.apply(lambda x: x.str.replace(',', '.'))
df_punto_coma
```

```python
df_punto_coma['Precio m2 abr 2022'] = pd.to_numeric(df_punto_coma['Precio m2 abr 2022'],errors='coerce')
```

```python
df_ordenado = df_punto_coma.drop(['Variación mensual', 'Variación trimestral', 'Variación anual', 'Máximo histórico', 'Variación máximo'], axis = 1).sort_values(by=['Precio m2 abr 2022'], ascending = False)
df_ordenado
```

```python
los_que_si = ['Madrid', 'Barcelona', 'Málaga', 'Sevilla', 'Valencia/Valéncia', 'Zaragoza']
```

```python
df_los_si = df_ordenado[df_ordenado['Localización'].isin(los_que_si)]
df_los_si
```

```python
df_los_si['Precio m2 abr 2022'] = df_los_si['Precio m2 abr 2022'].astype(int)
df_los_si
```

```python
df_los_si.to_csv('informe_de_precios_en_alquiler_españa_limpio.csv', index = False)
```
