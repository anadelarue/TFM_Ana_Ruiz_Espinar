```python
df_sucio = pd.read_csv('hogares_por_regimen_de_tenencia_de_la_vivienda_y_edad_y_sexo_de_la_persona_de_referencia.csv', encoding = 'latin1', sep = ';')
```

```python
df_sucio
```

```python
periodos = [2020, 2007]
```

```python
df_años = df_sucio[df_sucio['Periodo'].isin(periodos)]
df_años
```

```python
df_años['Total'] = df_años['Total'].map(lambda x: x.replace(',', '.'))
df_años
```

```python
df_años['Total'] = df_años['Total'].astype(float)
df_años
```

```python
año = [2020]
```

```python
df_2020 = df_años[df_años['Periodo'].isin(año)]
```

```python
df_2020.to_csv('hogares_por_regimen_de_tenencia_de_la_vivienda_y_edad_y_sexo_de_la_persona_de_referencia_limpio_2020.csv', index = False)
```
