```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_rows", None)
```

```python
df_sucio = pd.read_csv('poblacion_por_relacion_de_parentesco_con_la_persona_de_referencia,_sexo_y_grupo_de_edad.csv', encoding = 'latin1', sep=';')
```

```python
df_sucio
```

```python
años = [2007, 2020]
```

```python
df_años = df_sucio[df_sucio['Periodo'].isin(años)]
df_años
```

```python
df_años['Total'] = df_años['Total'].str.replace('.','')
df_años
```

```python
df_años['Total'] = df_años['Total'].map(lambda x : x.replace(',','.')).astype(float)
df_años
```

```python
df_agrupado = df_años.groupby(by = ['Periodo', 'Sexo'], as_index = False).median()
df_agrupado
```

```python
df_agrupado['Total'] = df_agrupado['Total'] * 100
df_agrupado
```

```python
df_agrupado['Total'] = df_agrupado['Total'].astype(int)
df_agrupado
```
