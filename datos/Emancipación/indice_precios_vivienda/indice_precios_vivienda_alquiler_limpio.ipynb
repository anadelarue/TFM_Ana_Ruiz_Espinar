```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
```

```python
df_sucio = pd.read_csv('indice_precios_vivienda.csv')
```

```python
df_sucio
```

```python
datos_alquiler = {"Municipios": df_sucio['NMUN'], "Precio/m2": df_sucio['Renta_Medi'], "Superficie": df_sucio['Superficie']}
```

```python
df_alquiler = pd.DataFrame(datos_alquiler).dropna()
df_alquiler
```

```python
df_int = df_alquiler.astype({"Superficie":'int'})
df_int
```

```python
df_no_ceros = df_int[(df_int != 0).all(1)]
df_no_ceros
```

```python
df_casi_limpio = df_no_ceros.groupby(by=['Municipios'], as_index=False).median()
```

```python
df_casi_limpio
```

```python
los_que_si = ['Madrid', 'Barcelona', 'Málaga', 'Sevilla', 'Valencia', 'Zaragoza']
```

```python
df_los_si = df_casi_limpio[df_casi_limpio['Municipios'].isin(los_que_si)]
```

```python
df_los_si['Precio/m2'] = df_los_si['Precio/m2'].astype(int)
df_los_si
```

```python
df_los_si.to_csv('indice_precios_vivienda_alquiler_limpio.csv', index = False)
```
