```python
import pandas as pd
import numpy as np
```

```python
df_sucio = pd.read_csv('tasa_de_empleo_de_las_personas_de_25_a_49_anos_sin_hijos-con_hijos_menores_de_12_anos_segun_periodo.csv', encoding = 'latin1', sep = ';')
```

```python
df_sucio
```

```python
años = [2020]
```

```python
df_sucio = df_sucio[df_sucio['Periodo'].isin(años)]
```

```python
df_sucio
```

```python
df_sucio['Total'] = df_sucio['Total'].map(lambda x: x.replace(',', '.')).astype(float)
```

```python
df_sucio_hombres = df_sucio[0:5]
df_sucio_hombres.rename(columns = {'Total': 'Total hombres'}, inplace=True)
df_sucio_hombres.reset_index(drop=True, inplace=True)
```

```python
df_sucio_mujeres = df_sucio[5:]
df_sucio_mujeres.rename(columns = {'Total': 'Total mujeres'}, inplace=True)
df_sucio_mujeres.reset_index(drop=True, inplace=True)
```

```python
d = pd.concat([df_sucio_hombres, df_sucio_mujeres], axis=1, ignore_index=False)
```

```python
d
```

```python
d.insert (8, "Diferencia", d['Total hombres'] - d['Total mujeres'])
```
