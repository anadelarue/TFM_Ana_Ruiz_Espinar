```python
df_sucio = pd.read_csv('hombres_cuyas_parejas_no_estan_embarazadas,_no_tienen_hijos_y_no_desean_tenerlos_segun_el_motivo_mas_importante_de_este_hecho_y_edad.csv', encoding = 'latin1', sep = ';')
```

```python
#Para poder hacerlo en R
df_sucio.to_csv('hombres_cuyas_parejas_no_estan_embarazadas,_no_tienen_hijos_y_no_desean_tenerlos_segun_el_motivo_mas_importante_de_este_hecho_y_edad_limpio.csv', index = False)
```

```python
df_sucio = df_sucio[0:35]
```

```python
df_sucio
```

```python
df_sucio['Total'] = df_sucio['Total'].map(lambda x: x.replace(',', '.')).astype(float)
```

```python
df_sucio_menos30 = df_sucio[0:7]
df_sucio_menos30.rename(columns = {'Total': 'Total < 30'}, inplace=True)
df_sucio_menos30.reset_index(drop=True, inplace=True)
```

```python
df_sucio_30_34 = df_sucio[7:14]
df_sucio_30_34.rename(columns = {'Total': 'Total 30 - 34'}, inplace=True)
df_sucio_30_34.reset_index(drop=True, inplace=True)
```

```python
df_sucio_35_39 = df_sucio[14:21]
df_sucio_35_39.rename(columns = {'Total': 'Total 35 - 39'}, inplace=True)
df_sucio_35_39.reset_index(drop=True, inplace=True)
```

```python
df_sucio_40_44 = df_sucio[21:28]
df_sucio_40_44.rename(columns = {'Total': 'Total 40 - 44'}, inplace=True)
df_sucio_40_44.reset_index(drop=True, inplace=True)
```

```python
df_sucio_45_49 = df_sucio[28:]
df_sucio_45_49.rename(columns = {'Total': 'Total 45 - 49'}, inplace=True)
df_sucio_45_49.reset_index(drop=True, inplace=True)
```

```python
d = pd.concat([df_sucio_menos30, df_sucio_30_34, df_sucio_35_39, df_sucio_40_44, df_sucio_45_49], axis=1, ignore_index=False)
```

```python
d
```

```python
d.insert (6, "Porcentaje 30 - 34", d['Total < 30'] - d['Total 30 - 34'])
```

```python
d.insert (10, "Porcentaje 35 - 39", d['Total 30 - 34'] - d['Total 35 - 39'])
```

```python
d.insert (14, "Porcentaje 40 - 44", d['Total 35 - 39'] - d['Total 40 - 44'])
```

```python
d.insert (18, "Porcentaje 45 - 49", d['Total 40 - 44'] - d['Total 45 - 49'])
```
